```python
import pandas as pd
```

```python
pd.options.display.float_format = '{:,.2f}'.format
    
def format_float(value):
    return f'{value:,.2f}'
pd.options.display.float_format = format_float
```

```python
df = pd.read_csv('Production.csv')
#Meses a días
days=30.28
#Metros cubicos a barriles (aceite)
bbls = 6.29
#Metros cubicos a pies cubicos (gas)
ft3 = 35.3147
#Millones
MM=1000000
df
```

```python
#coords = df[['POZO', 'LAT', 'LON']]
#coords = coords.dropna()
#coords.to_csv('Coords.csv')
```

```python
prod = df[['POZO', 'FECHA', 'DIAS', 'ACEITE', 'GAS', 'AGUA']]
prod['aceite_bpm'] = prod['ACEITE']*bbls
prod['aceite_bpd'] = prod['aceite_bpm']/prod['DIAS']
prod['agua_bpm'] = prod['AGUA']*bbls
prod['agua_bpd'] = prod['agua_bpm']/prod['DIAS']
prod['gas_cfpm'] = prod['GAS']*ft3
prod['gas_cfpd'] = prod['gas_cfpm']/prod['DIAS']
prod['gas_mmcfpm'] = prod['gas_cfpm']*MM
prod['gas_mmcfpd'] = prod['gas_cfpd']*MM
prod['FECHA'] = pd.to_datetime(prod['FECHA'])
prod['POZO'] = prod['POZO'].str.replace('_',' ')
prod['POZO'] = prod['POZO'].str.replace('GDE','GRANDE')
```

```python
#SOURCE - http://cdn.ihs.com/fekete/help/FieldNotes/c-te-calcexamples.htm
ratios = df[['POZO', 'FECHA', 'ACEITE', 'GAS', 'AGUA']]
ratios['WOR'] = ratios['AGUA']/ratios['ACEITE']
ratios['GOR'] = ratios['GAS']/ratios['ACEITE']
ratios['WC'] = ratios['AGUA']/(ratios['AGUA']+ratios['ACEITE'])
ratios = ratios.loc[:, ~ratios.columns.str.contains('^Unnamed')]
ratios.columns = [x.lower() for x in ratios.columns]
ratios['pozo'] = ratios['pozo'].str.replace('_',' ')
ratios['pozo'] = ratios['pozo'].str.replace('GDE','GRANDE')
ratios['fecha'] = pd.to_datetime(ratios['fecha'])
#ratios.to_csv('Ratios.csv', index=False)
ratios
```

```python
df3 = pd.read_csv("RPFC-Plano de referencia.csv")
df3.columns = [x.lower() for x in df3.columns]
df3 = df3.rename(columns={"terminacion": "pozo"})
df3['fecha'] = pd.to_datetime(df3['fecha'])
df3['pozo'] = df3['pozo'].str.replace('SG-','SITIO GRANDE ')
df3['tipo de registro'] = df3['tipo de registro'].str.replace('C','CERRADO')
df3['estrangulador (pg)'] = df3['estrangulador (pg)'].fillna('N/A')
df3 = df3.fillna(0)
press_pref = df3.copy()
press_pref.to_csv('RPFC-Plano de referencia.csv', index=False)
```
